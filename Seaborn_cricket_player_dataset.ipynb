```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
# Load the dataset
df = pd.read_csv('/content/cricket_players.csv')

# Display the first 5 rows
print(df.head())
```

```python
plt.figure(figsize=(10, 6))
sns.barplot(data=df, x='Country', y='Runs', estimator=sum, ci=None)
plt.title('Total Runs by Country')
plt.xlabel('Country')
plt.ylabel('Total Runs')
plt.grid(True)
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(10, 6))
sns.boxplot(data=df, x='Country', y='Average')
plt.title('Batting Average Distribution by Country')
plt.xlabel('Country')
plt.ylabel('Batting Average')
plt.grid(True)
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(10, 6))
sns.barplot(data=df, x='Country', y='Runs', estimator=sum, palette='Set2', ci=None)
plt.title('Total Runs by Country - Customized')
plt.xlabel('Country')
plt.ylabel('Total Runs')
plt.grid(True)
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(10, 6))
sns.scatterplot(data=df, x='Matches Played', y='Runs', hue='Country', palette='Set2')
plt.title('Runs vs Matches Played')
plt.xlabel('Matches Played')
plt.ylabel('Runs')
plt.grid(True)
plt.tight_layout()
plt.show()
```
